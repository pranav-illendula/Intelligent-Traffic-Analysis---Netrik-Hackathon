```python
from ultralytics import YOLO
import torch

# Check GPU
device = 0 if torch.cuda.is_available() else "cpu"
print("Using device:", device)

# Load pretrained base model
model = YOLO("yolov8s.pt")   # best balance of speed + accuracy

# Start training
model.train(
    data="/content/Auto-Rickshaw--1/data.yaml",  # path to your yaml
    epochs=120,
    imgsz=960,
    batch=8,
    device=device,
    workers=4,

    # ğŸ”¥ AUGMENTATIONS (important)
    hsv_h=0.015,
    hsv_s=0.7,
    hsv_v=0.4,
    degrees=10,
    translate=0.1,
    scale=0.5,
    shear=2,
    fliplr=0.5,
    mosaic=1.0,
    mixup=0.2,

    # Regularization
    patience=25,
    cos_lr=True
)
```
